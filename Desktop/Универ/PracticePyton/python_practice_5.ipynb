5. (11б). Створіть 2 дочірніх класи від Фігури: Трикутник та Коло, які мають атрибути, необхідні для розрахунку площин. Визначте метод get_area в кожному з них так, щоби вони розраховували площу в залежності від формули для кожного типу фігури. Створіть обʼєкт класу Трикутник та обʼєкт класу Коло. Виконайте операцію суми за допомогою оператора + між ними.

```python
import math

class Figure:
    def get_area(self):
        return 0

    def __add__(self, other):
        return self.get_area() + other.get_area()

class Triangle(Figure):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def get_area(self):
        return 0.5 * self.base * self.height

class Circle(Figure):
    def __init__(self, radius):
        self.radius = radius

    def get_area(self):
        return math.pi * self.radius**2
    
    
    
triangle = Triangle(base=5, height=8)
circle = Circle(radius=4)


print(f"Area of Triangle: {triangle.get_area()}")
print(f"Area of Circle: {circle.get_area()}")

total_area = triangle + circle
print(f"Total Area: {total_area}")
```

7. (16б). Створіть клас BankAccount з приватними атрибутами balance та account_number.
Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори getter та setter.

```python
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self._account_number = account_number
        self._balance = initial_balance

    @property
    def balance(self):
        return self._balance

    @balance.setter
    def balance(self, new_balance):
        if new_balance < 0:
            print("Баланс не може бути від'ємним.")
        else:
            self._balance = new_balance
            print(f"Баланс оновлено: {self._balance}")

    def deposit(self, amount):
        """Метод для поповнення рахунку."""
        if amount > 0:
            self.balance += amount
            print(f"Поповнення на {amount} грн успішно.")
        else:
            print("Сума поповнення повинна бути більше 0.")

    def withdraw(self, amount):
        """Метод для зняття грошей з рахунку."""
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Зняття {amount} грн успішно.")
        elif amount <= 0:
            print("Сума зняття повинна бути більше 0.")
        else:
            print("Недостатньо коштів на рахунку.")

# Приклад використання класу BankAccount
account1 = BankAccount(account_number="1")
print(f"Початковий баланс: {account1.balance} грн")
account1.deposit(500)
account1.withdraw(300)

```

8. (16б). Створіть клас Library, який містить список об'єктів типу Book.
Реалізуйте методи для додавання книги, видалення книги та відображення списку книг.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        #"""Метод для додавання книги до бібліотеки."""
        new_book = Book(title, author)
        self.books.append(new_book)
        print(f"Книга '{title}' додана до бібліотеки.")

    def remove_book(self, title):
       # """Метод для видалення книги з бібліотеки за назвою."""
        for book in self.books:
            if book.title == title:
                self.books.remove(book)
                print(f"Книга '{title}' видалена з бібліотеки.")
                return
        print(f"Книга '{title}' не знайдена в бібліотеці.")

    def display_books(self):
       # """Метод для відображення списку книг у бібліотеці."""
        if not self.books:
            print("У бібліотеці немає книг.")
        else:
            print("Список книг у бібліотеці:")
            for book in self.books:
                print(f"- {book.title} (Автор: {book.author})")

# Приклад використання класу Library
library = Library()

library.add_book("Гаррі Поттер і філософський камінь", "Дж.К. Роулінг")
library.add_book("Три товариші", "Е. М. Ремарк")


library.display_books()

library.remove_book("Три товариші")

library.display_books()
```

9. (13б). Створіть клас Person з атрибутами name та age.
Створіть ще один клас Employee з такими атрибутами, як department та salary.
Створіть клас Manager, який успадковує обидва класи Person та Employee. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Manager та отримавши mro для цього класу.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Employee:
    def __init__(self, department, salary):
        self.department = department
        self.salary = salary

class Manager(Person, Employee):
    def __init__(self, name, age, department, salary):
        # Викликаємо конструктори батьківських класів
        Person.__init__(self, name, age)
        Employee.__init__(self, department, salary)
        
        
        
manager = Manager(name="John Doe", age=35, department="HR", salary=50000)

# Виведення атрибутів об'єкта
print(f"Name: {manager.name}")
print(f"Age: {manager.age}")
print(f"Department: {manager.department}")
print(f"Salary: {manager.salary}")


print(f"MRO: {Manager.mro()}")
```
